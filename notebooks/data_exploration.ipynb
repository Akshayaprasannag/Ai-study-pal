```python
# =============================================
# AI Study Pal - Phase 1.2 : Dataset Preparation & Visualization
# =============================================

import pandas as pd
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Create a Small Sample Dataset
data = {
    "subject": [
        "Math", "Science", "History", "English", "Computer Science",
        "Math", "Science", "History", "English", "Computer Science"
    ],
    "text": [
        "Mathematics is the study of numbers, shapes, and logical patterns.",
        "Science explores the natural world through experiments and observation.",
        "History involves studying past events to understand human civilization.",
        "English focuses on language, grammar, and effective communication.",
        "Computer Science deals with computation, algorithms, and programming.",
        "Mathematics improves problem-solving and analytical thinking.",
        "Science enhances curiosity and innovation through research.",
        "History teaches lessons from ancient and modern times.",
        "English helps express thoughts clearly through reading and writing.",
        "Computer Science powers modern technology and artificial intelligence."
    ]
}

df = pd.DataFrame(data)
print("üîπ Original Dataset:")
print(df.head())

# 2Ô∏è‚É£ Basic Cleaning
# - Remove duplicates
# - Convert text to lowercase

df['text'] = df['text'].str.lower()
df = df.drop_duplicates()

print("\nüîπ Cleaned Dataset:")
print(df.head())

# 3Ô∏è‚É£ Save the cleaned dataset
df.to_csv("../data/cleaned_dataset.csv", index=False)
print("\n‚úÖ Cleaned dataset saved as 'data/cleaned_dataset.csv'")

# 4Ô∏è‚É£ Simple EDA
print("\nüîπ Subject Value Counts:")
print(df['subject'].value_counts())

# 5Ô∏è‚É£ Visualization
plt.figure(figsize=(6,6))
df['subject'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    colors=plt.cm.Paired.colors
)
plt.title("Subject Distribution in Dataset")
plt.ylabel("")  # Hide y-axis label
plt.show()
```
